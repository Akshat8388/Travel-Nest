### Columns Details:

1. **Zone**: Geographical region of the place within India, categorizing it into zones like Northern, Southern, etc.
2. **State**: The state in which the place is located, providing a regional context.
3. **City**: The city or town where the destination is situated.
4. **Name**: The name of the tourist spot or landmark.
5. **Type**: Classification of the place, such as Temple, War Memorial, or Natural Park.
6. **Establishment Year**: The year in which the place was established or discovered.

9. **Entrance Fee in INR**: The cost of admission in Indian Rupees.
10. **Airport within 50km Radius**: Indicates if there is an airport within 50 kilometers of the place for accessibility.
11. **Weekly Off**: The day of the week when the place is closed to visitors.
12. **Significance**: The importance or role of the place, such as Historical, Religious, or Environmental.
13. **DSLR Allowed**: Indicates whether visitors are permitted to use DSLR cameras at the location.
14. **Number of Google Reviews (in lakhs)**: The total number of Google reviews in lakhs (hundreds of thousands) that the place has received.
15. **Best Time to Visit**: Suggested time of the day for visiting the place to have the best experience.

## Building the First Recommendations Systems using Weighted Average Ratings

- This system suggests items (places) based on their average ratings and the number of reviews they have received. It uses a weighted average formula to balance the importance of ratings and review counts.
- How It Works:
Each place is assigned a score based on its average rating and the number of reviews.


- The score can be normalized to ensure all scores fall within a similar range.
- Outcome: It highlights popular places with high ratings, ensuring that places with many reviews are considered more reliable.


- Formula = Rv + Cm/ v + m 
- R is the average rating of the place

- v is the number of ratings for the place

- m is a threshold for the minimum number of reviews required

- C is the mean rating across all places

```python
first_recommendations_data = data
```

```python
m =1
C = first_recommendations_data['Google review rating'].mean()

def weighted_rating(x, m=m, C=C):
    v = x['Number of google review in lakhs']  
    R = x['Google review rating']
    return (v / (v + m) * R) + (m / (v + m) * C)


first_recommendations_data['Weighted Rating'] = first_recommendations_data.apply(weighted_rating, axis=1)
top_places = first_recommendations_data.sort_values('Weighted Rating', ascending=False)
top_places.head(10)
```

### Recommendation based on scaled weighted avg and number of reviews (Priority is given 50% to both)

```python
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
scaling = mm.fit_transform(top_places[["Weighted Rating","Number of google review in lakhs"]])
scaling_data = pd.DataFrame(scaling,columns=["Weighted Rating_new","Number of google review in lakhs_new"])
scaling_data.head()
```

```python
top_places[["Weighted Rating_normalize","Number of google review in lakhs_normalize"]] = scaling_data
```

```python
top_places["recommendations"] = top_places['Weighted Rating_normalize'] *0.5 + top_places['Number of google review in lakhs_normalize']*0.5
top_places = top_places.sort_values(by="recommendations",ascending=False)
top_places.head(30).sample(6)
```

## Building Second Recommendations systems with content-based filtering

- This system recommends items based on the characteristics and features of the items themselves. It compares the attributes of a selected place with others to find similar ones.
- How It Works:
- Features of places (like type, significance, location) are transformed into a numerical format (e.g., using one-hot encoding).
- Cosine similarity or other similarity measures are calculated between the feature vectors of different places.
- When a user selects a place, the system finds and recommends other places with the highest similarity scores.
- Outcome: It provides personalized recommendations based on the features of places the user has shown interest in, allowing users to discover similar attractions.

```python
second_recommendations_data = first_recommendations_data
second_recommendations_data.drop(columns=["Weighted Rating","time needed to visit in hrs"],inplace=True)
```

```python
second_recommendations_data.head()
```

```python
features = ['Zone', 'State', 'City', 'Type','Google review rating', 'Significance', 'Best Time to visit']
```

```python
features_matrix = second_recommendations_data[features]
```

```python
features_matrix_encoded = pd.get_dummies(features_matrix, columns=['Zone', 'State', 'City', 'Type', 'Significance', 'Best Time to visit'],dtype=int,drop_first=True)
```

```python
features_matrix_encoded
```

```python
features_matrix_encoded["Google review rating"]= mm.fit_transform(features_matrix_encoded[["Google review rating"]])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
#Cosine similarity is a metric used to measure how similar two vectors are, regardless of their size. 
#It’s particularly useful in high-dimensional spaces, such as in text analysis, recommendation systems
#, and machine learning, where items are often represented as vectors of features.
```

```python
cos_sim = cosine_similarity(features_matrix_encoded)
```

```python
def get_recommendations(place_name,second_recommendations_data,cos_sim):
    index = second_recommendations_data[second_recommendations_data["Name"]==place_name].index[0]
    similarity_scores = list(enumerate(cos_sim[index]))
    similarity_scores = sorted(similarity_scores,key=lambda x:x[1],reverse=True)
    similarity_scores = similarity_scores[1:7]
    place_indices = [i[0] for i in similarity_scores]
    return second_recommendations_data.iloc[place_indices][["Name","State"]]
```

```python
recommended_places = get_recommendations('Golden Temple (Harmandir Sahib)',second_recommendations_data,cos_sim)
recommended_places
```

## Third Recommendations Systems with Regional Data

```python
third_recommendations_data = data[["Zone","State","City","Name","Google review rating","Significance","Best Time to visit","Image URL"]]
```

```python
third_recommendations_data.head()
```

```python
region = data["Zone"]
X = third_recommendations_data[["State","City","Name","Google review rating","Significance","Best Time to visit"]]
```

```python
print(region.unique())
```

```python
X_encoded = pd.get_dummies(X,columns=["State","Name","City","Significance","Best Time to visit"],dtype=int,drop_first=True)
```

```python
X_encoded = mm.fit_transform(X_encoded)
```

```python
from sklearn.cluster import KMeans
```

```python
# for finding the best value of k 
wcss = []
for i in range(1,11):
    km = KMeans(n_clusters=i,random_state=1)
    km.fit(X_encoded)
    wcss.append(km.inertia_)
print(wcss)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(1,11), wcss, linestyle='-', color='b')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')
plt.grid(True)
```

```python
km = KMeans(n_clusters=6,random_state=42)
km.fit(X_encoded)
```

```python
y = km.predict(X_encoded)
y
```

```python
third_recommendations_data_copy = third_recommendations_data
```

```python
third_recommendations_data_copy["Label"] = y
```

```python
cluster_counts = third_recommendations_data_copy.groupby(['Zone', 'Label']).size().unstack(fill_value=0)
print(cluster_counts)
```

```python
cluster_counts.plot(kind='bar', stacked=True, figsize=(12, 8))
plt.title('Cluster Distribution by Region')
plt.xlabel('Region')
plt.ylabel('Number of Places')
plt.xticks(rotation=45)
plt.legend(title='Cluster')
plt.show()
```

```python
r = third_recommendations_data_copy[third_recommendations_data_copy['Zone']=="Northern"]
```

```python
cluster = r['Label'].unique()
```

```python
rem = pd.DataFrame()
for c in cluster:
    l = r[r['Label']==c]
    top_destinations = l.nlargest(20, 'Google review rating')
    rem = pd.concat([rem, top_destinations], ignore_index=True)
```
